### Example of Kingman process times generation on the branch with `n>1` 
(lineages did not fully coalesced before the time `tau`)

```python
M = 100
theta = 1
tau = 0.1
times = efuns.generate_kingman_times(M, theta)
times = list(filter(lambda x: x<tau, times))
print(len(times))
```

Estimations in this case

```python
theta_exp = efuns.estimate_theta_exp(times, M=M, tau=tau)
theta_lln = efuns.estimate_theta_lln(times, M=M, tau=tau)
print(theta_exp, theta_lln)
```
